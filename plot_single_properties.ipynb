```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Carboxylic.csv', usecols=["HBA"])
plt.hist(df, range=(0, +140), bins=100, histtype='step', log=True, label="HBA",color='red')
plt.xlabel("HBA", size=14)
plt.ylabel('Frequency', size=14)
plt.legend(frameon=False, loc='upper right', ncol=2, title='Carboxylates')
plt.tight_layout() 
plt.savefig("carxoxylates_HBA.png", dpi=300, facecolor='white')
```
